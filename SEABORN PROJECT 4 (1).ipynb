```python
# 2x2 Grid

import seaborn as sns
sns.set(style = "ticks",color_codes = True)

tips = sns.load_dataset("tips")

g = sns.FacetGrid(tips, col = "time", row = "smoker")

```

```python
# create a univariate plot

import matplotlib.pyplot as plt

g = sns.FacetGrid(tips, col = "time",row = "smoker")

g= g.map(plt.hist, "total_bill")
```

```python
import numpy as np

bins = np.arange(0, 65, 5 )

g = sns.FacetGrid(tips, col = "time", row = "smoker")

g = g.map(plt.hist, "total_bill", bins = bins, color = "g")
```

```python
# bivariate


g = sns.FacetGrid(tips, col = "time", row = "smoker")

g = g.map(plt.scatter, "total_bill", "tip", edgecolor = "b")


```

```python

g = sns.FacetGrid(tips, col = "time", hue = "smoker")

g = g.map(plt.scatter, "total_bill", "tip", edgecolor = "y")


```

```python
g = sns.FacetGrid(tips, col="time", hue = "smoker")

g = g.map(plt.scatter, "total_bill", "tip", edgecolor= "y")

```

```python
# change the height and aspect ratio

g= sns.FacetGrid(tips, col= "day", height = 4, aspect = 0.4)
g=g.map(plt.hist, "total_bill", bins = bins)
```

```python
# Specify Order

g= sns.FacetGrid(tips, col= "smoker", col_order = ["Yes", "No"], height = 6, aspect = 1)
g=g.map(plt.hist, "total_bill", bins = bins, color = "m")

```

```python
# use a different color palette

kws = dict(s=50, linewidth= 0.5, edgecolor = "w")

g= sns.FacetGrid(tips, col = "sex", hue= "time", palette = "Set1",
                hue_order=["Dinner", "Lunch"], height = 5)

g=g.map(plt.scatter, "total_bill","tip" )
```

```python
#dictionary mapping for hue colors

pal = dict(Lunch="springgreen", Dinner= "grey")

g= sns.FacetGrid(tips, col = "sex", hue= "time", palette = pal,
                hue_order=["Dinner", "Lunch"], height = 5)

g=g.map(plt.scatter, "total_bill","tip" )
```

```python
# different axis labels after plotting

g = sns. FacetGrid(tips, col="smoker", row ="sex")

g = (g.map(plt.scatter,"total_bill","tip", color = "g", **kws)
        .set_axis_labels("Total bill(US Dollars)", "Tip"))

     

```

```python
#set other attributes taht are shared across the facets

g = sns. FacetGrid(tips, col="smoker", row ="sex")

g = (g.map(plt.scatter,"total_bill","tip", color = "r", **kws)
        .set(xlim = (0,60), ylim =(0,12),
             xticks = [10,30,50], yticks = [2,6,12]))
             
```

```python
#white space
#tighten the facets

g = sns. FacetGrid(tips, col="smoker", row ="sex",
                  margin_titles = True)

g = (g.map(plt.scatter,"total_bill","tip", color = "r", **kws)
        .set(xlim = (0,60), ylim =(0,12),
             xticks = [10,30,50], yticks = [2,6,12])
        .fig.subplots_adjust(wspace = .5, hspace = .5))
```

```python
#white space
#tighten the facets

g = sns. FacetGrid(tips, col="smoker", row ="sex",
                  margin_titles = True)

g = (g.map(plt.scatter,"total_bill","tip", color = "r", **kws)
        .set(xlim = (0,60), ylim =(0,12),
             xticks = [10,30,50], yticks = [2,6,12])
        .fig.subplots_adjust(wspace = .5, hspace = 1))
```

```python
# template for facet titles 
# change col wrap

g = sns.FacetGrid(tips, col = "size", col_wrap = 3)
g = (g.map(plt.hist, "tip", bins=np.arange(0,13), color = "y")
     .set_titles("{col_name} diners"))
     
     
```

```python
# different marker for the hue levels

g=sns.FacetGrid(tips, col = "sex", hue= "time", palette = pal,
               hue_order = ["Dinner", "Lunch"],
                hue_kws = dict(marker = ["+", "v"]), height = 6, aspect = 1)

g= (g.map(plt.scatter, "total_bill", "tip", **kws)
    .add_legend())


                
```

```python
# wrap a column into many levels into the rows

att= sns.load_dataset("attention")

g=sns.FacetGrid(att, col= "subject", col_wrap = 6, height = 1.5)

g=g.map(plt.plot,"solutions", "score", marker = "." )
```
